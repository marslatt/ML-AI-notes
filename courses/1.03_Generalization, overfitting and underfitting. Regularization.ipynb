### Uderfitting, generalization, overfitting

#### Unerfitting
* If the model does not fit the training data very well, it is said that the model underfits the data.  
* Another term for underfitting is that the model has <b>high bias</b>. 

#### Generalization
* If the lmodel does well even on examples, that are not on the training set, that's called <b>generalization</b>. 
* We want the model to generalize well, which means to make much more reasonable predictions even on brand new examples that it has never seen before.  
* Generalization example with binary classification:
  * Including most samples of one class within a single decision boundary is considered a pretty good fit to the data, even though the boudary may not perfectly classify every single training example in the training set and some of the samples of the other class may get classified among it too. 
  * The model is considered to show pretty good overal performance (i.e. just right) if it generalizes well enough to new samples.
  * If the model tries really hard and contours or twist itself to find a decision boundary that fits the training data perfectly, it will produces many higher-order polynomial features with a very complex decision boundary, and will likely overfit the training data.

#### Overfitting
* If the model is doing extremely good job fitting the training data, but make poorly predictions on examples, that are not on the training set, the algorith is overfitting the training dataset.
* Choosing such parameters, that will result in the cost function being $â‰ˆ0$ (because the errors on all training examples are $â‰ˆ0$), may lead to overfitting problem.
* Another term for the case where the model fits the training data almost too well, but not generalize well to new examples that's never seen before, is that the model has <b>high variance</b>.
* Simpler models are less likely to overfit, but more prone to overfitting (if too simple).

### Regularization

* One way to shrink certain model parameters $ð‘¤$ is by adding a penalty term to the cost function â€” this is called regularization. The strength of the penalty is controlled by a coefficient $\lambda$ (lambda).
* More generally, the regularizatin tend to be implemented with penalizing all parameters $(w_{1},...,w_{n})$, since it's not known in advance which are the most important features and which are to penalize. 
* The task of the Gradient descent algorithm is then to find really small 'penalized' values of $w$ to be able to minimize the modified cost, otherwise they will weigh too much in the final sum.
* Regularization usually rezults in fitting a smoother and simpler model, which is less likely to overfit.
* The regularization parameter $\lambda$ determines how much large weights are discouraged:
  * If $\lambda$ = 0 &rarr; we don't penalize anything &rarr; possible overfitting
  * If $\lambda$ >>> 0 (very large) &rarr; too heavy regularization weights &rarr; $w_{1}...w_{n}$ will be very small after learning &rarr; possible underfitting
  * Increase in $\lambda$ leads to decrease in $w_{1}...w_{n}$ &rarr; optimize!
* Types of cost function with regularization: 
<br>&emsp;&emsp;&emsp;&nbsp;&nbsp; <b>L2</b> Regularization (Ridge): $J(\vec{w},b)_{regularized} = J(\vec{w},b) + \sum_{j=1}^{n}w_{j}^2$
<br>
<br>&emsp;&emsp;&emsp;&nbsp;&nbsp; <b>L1</b> Regularization (Lasso): $J(\vec{w},b)_{regularized} = J(\vec{w},b) + \sum_{j=1}^{n}|w_{j}|$

### How to solve underfitting and overfitting?

#### Overfitting
* Use <b>regularization</b> - this is a way to more gently reduce the impact of some of the features, without eliminating them all together. (I)
  * Regularization encourages the model to shrink the parameters, without neccessarily demanding to set them to 0.
  * Usually $w$ params are regularized, not $b$.
* Collect more trainign data. (II)
* If there's not enough training samples &rarr; reduce the number of features used i.e. "feature selection". (III)
  * Simpler models are less likely to overfit.
  * The side effect of feature selection is that we throw away some useful information.
  * There are even models that automatically chose most appropriate features.


#### Underfitting: 
* Use a more complex model - add better features such as polynomial features 
* Train the model longer.
* Reduce regularization (lower lambda).
