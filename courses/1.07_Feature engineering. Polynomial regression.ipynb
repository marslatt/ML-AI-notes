## Fefature engineering

> Feature engineering is the pre-processing step of machine learning, which extracts features from raw data.
>  <br><br> 
> More on feature engineering:
> * https://www.tpointtech.com/feature-engineering-for-machine-learning
> * https://www.ibm.com/think/topics/feature-engineering

* In regards to polynomial regression: feature engineering means using intuition to design new features, by transforming and combining original features.
* This allows us to use the machinery of linear regression to fit very complicated and non-linear curves to our data.
* Example for linear regression model: $f_{\vec{w},b}(\vec{x})$ = $w_{1}x_{1}$ + $w_{2}x_{2}$ + $b$ 
<br>&emsp;&emsp;&emsp;&nbsp;&nbsp; regular features &rarr; [ $x_{1}$ &nbsp; $x_{2}$ ]
<br>&emsp;&emsp;&emsp;&nbsp;&nbsp; polynomial features &rarr;  [ $x_{1}$ &nbsp; $x_{2}$ &nbsp; $x_{1}^{2}$ &nbsp; $x_{2}^{2}$ &nbsp;  $x_{1}x_{2}$ ]


## Polynomial regression 

> "In polynomial regression, the original features are converted into polynomial features of required degree (2,3,..,n) and then modeled using a linear model."
> 
> It's a form of regression analysis in which the relationship between an independent variable x and a quantitative dependent variable y is modeled as an nth-degree polynomial.

* Model representation: $f_{\vec{w},b}(x)$ = $w_{1}x_{1}$ + $w_{2}x_{2}$ + $w_{3}x_{1}^{2}$ + $w_{4}x_{2}^{2}$ + $w_{5}x_{1}x_{2}$ + ... + $b$  
* * Based on linear regression model: $f_{\vec{w},b}(x)$ = $w_{1}x_{1}$ + $w_{2}x_{2}$ + ... + $b$
* Relies on feature engineering and the machinery of linear regression (i.e. it's a linear model in terms of parameters) to fit complicated non-linear functions. 
* If you're using gradient descent, it's important to apply feature scaling to get your features into comparable ranges of values. 
* * Since the values grow exponentialy, we want to keep them small from the beginning to make gradient descent work faster.
* Note that we can also use $\sqrt{x}$, $x^{-\frac{1}{3}}$ and etc., not only powers greater that 0.
* * In this way the curve will never flatten, but will be much less steep.
* When training, whatch out what parameters gradient descent is producing, especially if you are not sure what polynom would fit best.
* * Gradient descent will emphasize the coefficients that helps most to fit the data. The longer you run the training - the more significant they will become, compared to the rest.
* * Example: If you add $x^{3}$ compnonent to a model that needs only $x^{2}$ &rarr; gradient descent will produce a very small coefficient for $x^{3}$
